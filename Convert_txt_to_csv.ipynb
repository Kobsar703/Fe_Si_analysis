```python
import numpy as np
import pandas as pd
```

```python
def rewrite_no_char(file_name, file_name_convert):
    with open(file_name, 'r') as infile, open(file_name_convert, 'w') as outfile:
        # read file line by line
        for line in infile:
            # check if line starts with #
            if not line.startswith('#'):
                # if line doesn't start with #, write it to output file
                outfile.write(line)
            else:
                outfile.write(line[1:])
```

```python
rewrite_no_char("Abun_Fe_magn.dat", "Abun_Fe_magn_convert.dat")
```

```python
def convert_col(data, num, name1, name2):
    data[num] = data[num].str.replace('#', '')
    data[[name1, name2]] = data[num].str.split(expand=True).astype(float)
    data.drop(num, axis=1, inplace=True)
    return data
```
